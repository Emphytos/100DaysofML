## Line graphs
It seems that is the simplest graph in the library.

````python
rank = data['Rank']
score = data['Score']
````

````python
plt.plot(rank, score)
plt.show()
````

Without any axis-label it could be difficult to understand how look at this graph.

Here on the x-axis we can see the ranking and in the y-axis the score. Naturally, the higher the score is the better the rank is.

````python
plt.plot(rank, score)
plt.xlabel("Country rank")
plt.ylabel("Country score")
plt.title("Happiness Score vs World rank")
plt.show()
````

## Scatter plots

Scatter plots are a good way to visualize relationship between 2 variables.

````python
plt.scatter(gdp, score)
plt.title('GDP vs Happiness Score')
plt.xlabel('GDP per Capita')
plt.ylabel('Happiness Score')
plt.show()
````

It is not surprising that there is a correlation between both variables.

By default the graph starts from 0, this graph has the lowest y-label as 3. 

We can fix it by adding ```plt.ylim(0,8)```

````python
plt.scatter(gdp, score)
plt.title('GDP vs Happiness Score')
plt.xlabel('GDP per Capita')
plt.ylabel('Happiness Score')
plt.ylim(0,10)
plt.show()
````

For 0 score of GDP per Capita we've different Happiness score. That means that there are others factors than GDP per Capita

## Histograms

A historam shows a distribution of a particular of the data

````python
plt.hist(score)
plt.title('Happiness Score Distribution')
plt.xlabel('Happiness Score')
plt.ylabel('Count')
plt.show()
````

## Bar graphs
It is a bit harder to build them

````python
roundedHappinessScore = score.apply(int)
count = roundedHappinessScore.value_counts()
hapScore = count.index

plt.bar(hapScore, count)
plt.title('Happiness Scores')
plt.xlabel('Score')
plt.ylabel('Count')
plt.show()
````

The first line converts all of the happiness scores to integer values.

The second line count the number of times each score occurs

The third line gets the score associated with each count (needed as the x-axis of our graph)
